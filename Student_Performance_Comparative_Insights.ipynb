```python
import pandas as pd

# Load the dataset
file_path = r"C:\Users\laptop\Desktop\school_management_system\students\exported_students_data.csv"
data = pd.read_csv(file_path)

# Define the variables for comparison
variables = [
    'Academic Stress', 'Motivation', 'Sleep Disorder', 'Depression',
    'Has Electronic Device', 'Has Social Media Accounts',
    'Social Media Impact On Studies'
]

# Initialize a DataFrame to store the results
comparative_analysis_2 = pd.DataFrame()

# Loop over each variable and compute crosstab percentages
for var in variables:
    result = pd.crosstab(data[var], data['Academic Performance'], normalize='columns') * 100
    comparative_analysis_2 = pd.concat([comparative_analysis_2, result], axis=0)

# Display the results
print(comparative_analysis_2.round(2))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# تحميل البيانات
file_path = r"C:\Users\laptop\Desktop\school_management_system\students\exported_students_data.csv"
df = pd.read_csv(file_path)

# حساب المتوسطات حسب الأداء الأكاديمي
aggregated_means = df.groupby('Academic Performance')[[
    'Attendance Percentage', 'Daily Study Hours', 'Daily Gaming Hours'
]].mean().reset_index()

# ضبط الرسم
aggregated_means.set_index('Academic Performance').plot(kind='bar', figsize=(10, 6))

# تخصيص العناوين والمحاور
plt.title('Average Attendance, Study, and Gaming Hours by Academic Performance')
plt.ylabel('Mean Values')
plt.xlabel('Academic Performance Category')
plt.xticks(rotation=0)
plt.legend(title='Indicators')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# عرض الرسم
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
file_path = r"C:\Users\laptop\Desktop\school_management_system\students\exported_students_data.csv"
df = pd.read_csv(file_path)

# Variables to analyze
variables = ['Academic Stress', 'Motivation', 'Sleep Disorder', 'Depression']

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 8))
axes = axes.flatten()

for i, var in enumerate(variables):
    ct = pd.crosstab(df[var], df['Academic Performance'], normalize='columns') * 100
    ct.plot(kind='bar', ax=axes[i])
    axes[i].set_title(f'{var} by Academic Performance')
    axes[i].set_ylabel('Percentage')
    axes[i].legend(title='Performance')

plt.tight_layout()
# Save if needed
# plt.savefig("Figure_C20_CrosstabAnalysis.png", dpi=300)
plt.show()
```
